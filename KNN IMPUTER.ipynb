```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('train.csv', usecols=['Age', 'Survived','Pclass', 'Fare'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
mi = KNNImputer(n_neighbors=5)
```

```python
knn1 = mi.fit_transform(X_train)
knn2= mi.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(knn1,y_train)
y_pred = lr.predict(knn2)
accuracy_score(y_test,y_pred)
```

```python
si = SimpleImputer()
```

```python
trf1 = si.fit_transform(X_train)
trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(trf1,y_train)
y_pred = lr.predict(trf2)
accuracy_score(y_test,y_pred)
```
