```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Placement.csv')
df.head()
```

```python
# Boxplot for IQ
sns.boxplot(x=df['IQ'])
plt.title("IQ Outliers")
plt.show()

# Boxplot for CGPA
sns.boxplot(x=df['CGPA'])
plt.title("CGPA Outliers")
plt.show()
```

```python
mean_iq = df['IQ'].mean()
print("Mean value of IQ",mean_iq)
std_iq = df['IQ'].std()
print("Std value of IQ",std_iq)
```

```python
lower = mean_iq - 3 * std_iq
upper = mean_iq + 3 * std_iq
print("lower", lower)
print("Higher",upper)
```

```python
df.shape
```

```python
# Finding the outliers
no_of_outliers = df[(df['IQ'] < lower) | (df['IQ'] > upper)]
```

```python
no_of_outliers.shape
```

```python
# Trimming
new_df = df[(df['IQ'] < lower) & (df['IQ'] > upper)]
new_df.head()
```

```python
# Approach 2

# Calculating the Zscore

df['Iq_zscore'] = (df['IQ'] - mean_iq)/std_iq
```

```python
df[(df['Iq_zscore'] > 3) | (df['Iq_zscore'] < -3)]
```
