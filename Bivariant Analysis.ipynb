```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
titanic.head()
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# SCATTERPLOT(NUMERICAL _ NUMERICAL)
sns.scatterplot(x='total_bill', y='tip', data=df)
```

```python
sns.scatterplot(x='total_bill', y='tip',hue = df['sex'], style=df['smoker'], data=df) # multivariant
```

```python
# BARPLOT ( NUMERICAL TO CATEGORIAL)
sns.barplot(x='Pclass', y='Age', hue='Sex', data=titanic)
```

```python
#BOXPLOT( NUM TO CAT)
sns.boxplot(x='Sex', y='Age', data=titanic)
```

```python
#DISTPLOT (num - cat)
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

```python
#heatmap(cat - cat)
a = pd.crosstab(index=titanic['Pclass'],
                columns=titanic['Survived'],
                values=titanic['Age'],
                aggfunc='mean')
```

```python
sns.heatmap(a.corr())
```

```python
#pairplot
sns.pairplot(flights)
```

```python
import matplotlib.pyplot as plt
cols_multi = ['total_bill','tip','sex','smoker','day','time','size']
sns.pairplot(df[cols_multi])
plt.show()
```
