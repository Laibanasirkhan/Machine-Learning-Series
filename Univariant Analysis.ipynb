```python
mean_age = df['Age'].mean()
std_age = df['Age'].std()

# Define outlier thresholds
lower_limit = mean_age - 3 * std_age
upper_limit = mean_age + 3 * std_age

# Detect outliers
outliers = df[(df['Age'] < lower_limit) | (df['Age'] > upper_limit)]
```
